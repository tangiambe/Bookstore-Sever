```python
import pandas as pd
```

```python
book = pd.read_csv("book.csv")
bookauthor = pd.read_csv("bookauthor.csv")
```

```python
book = book.rename(columns={'id':'book_id'})
book.head()
```

```python
book.dtypes
```

```python
bookauthor
```

```python
bookauthor = bookauthor[bookauthor['book_id'].str.isnumeric()]
```

```python
bookauthor
```

```python
bookauthor.dtypes
```

```python
#convert book_id in the bookauthor df to int
bookauthor['book_id'] = bookauthor['book_id'].astype(int)
```

```python
bookauthor.dtypes
```

```python
#merging tables based on book_id
merged_df = pd.merge(book, bookauthor, on='book_id', how='inner') 
```

```python
columns_to_drop = ['_id_x', '_id_y', 'id']
merged_df = merged_df.drop(columns=columns_to_drop, axis=1)
merged_df.dropna()
```

```python
merged_df.dtypes
```

```python
merged_df = merged_df.rename(columns={'book_id':'id'})
```

```python
merged_df.to_csv("new_book.csv")
```
